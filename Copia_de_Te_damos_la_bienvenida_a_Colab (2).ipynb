```
"""
Sistema de Gestión de Vehículos
POO en Python - Encapsulamiento, Herencia y Polimorfismo
integrantes:
Jennifer Mazzini
Elian Andrade
"""


class Vehiculo:

    def __init__(self, marca: str, modelo: str, anio: int, kilometraje: float):
        self.__marca = marca
        self.__modelo = modelo
        self.__anio = anio
        self.__kilometraje = kilometraje

    @property
    def marca(self):
        return self.__marca

    @property
    def modelo(self):
        return self.__modelo

    @property
    def anio(self):
        return self.__anio

    @property
    def kilometraje(self):
        return self.__kilometraje

    @marca.setter
    def marca(self, valor: str):
        self.__marca = valor

    @modelo.setter
    def modelo(self, valor: str):
        self.__modelo = valor

    @anio.setter
    def anio(self, valor: int):
        self.__anio = valor

    @kilometraje.setter
    def kilometraje(self, valor: float):
        self.__kilometraje = valor

    def mostrar_informacion(self):
        print(f"  Marca      : {self.__marca}")
        print(f"  Modelo     : {self.__modelo}")
        print(f"  Año        : {self.__anio}")
        print(f"  Kilometraje: {self.__kilometraje:,.1f} km")

    def calcular_costo_mantenimiento(self) -> float:
        return 0.0


class Automovil(Vehiculo):
    VALOR_BASE = 50.0

    def __init__(self, marca, modelo, anio, kilometraje):
        super().__init__(marca, modelo, anio, kilometraje)

    def calcular_costo_mantenimiento(self) -> float:
        return self.VALOR_BASE + (self.kilometraje * 0.01)

    def mostrar_informacion(self):
        print(f"\n🚗 Automóvil")
        super().mostrar_informacion()
        print(f"  Costo mant.: ${self.calcular_costo_mantenimiento():,.2f}")


class Motocicleta(Vehiculo):
    VALOR_BASE = 35.0

    def __init__(self, marca, modelo, anio, kilometraje):
        super().__init__(marca, modelo, anio, kilometraje)

    def calcular_costo_mantenimiento(self) -> float:
        return self.VALOR_BASE + (self.kilometraje * 0.005)

    def mostrar_informacion(self):
        print(f"\n🏍️  Motocicleta")
        super().mostrar_informacion()
        print(f"  Costo mant.: ${self.calcular_costo_mantenimiento():,.2f}")


class Camioneta(Vehiculo):
    VALOR_BASE = 80.0

    def __init__(self, marca, modelo, anio, kilometraje):
        super().__init__(marca, modelo, anio, kilometraje)

    def calcular_costo_mantenimiento(self) -> float:
        return self.VALOR_BASE + (self.kilometraje * 0.02)

    def mostrar_informacion(self):
        print(f"\n🚙 Camioneta")
        super().mostrar_informacion()
        print(f"  Costo mant.: ${self.calcular_costo_mantenimiento():,.2f}")


# Función normal fuera de las clases
def sumar_total_mantenimiento(lista_vehiculos):
    total = 0.0
    for vehiculo in lista_vehiculos:
        total += vehiculo.calcular_costo_mantenimiento()
    return total


# Programa principal
auto1      = Automovil("Toyota",   "Corolla", 2020, 45000.0)
auto2      = Automovil("Honda",    "Civic",   2019, 62000.0)
moto1      = Motocicleta("Yamaha", "MT-07",   2022, 15000.0)
moto2      = Motocicleta("Honda",  "CB500F",  2021, 23000.0)
camioneta1 = Camioneta("Ford",     "F-150",   2018, 98000.0)
camioneta2 = Camioneta("Toyota",   "Hilux",   2023, 12000.0)

flota = [auto1, auto2, moto1, moto2, camioneta1, camioneta2]

print("=" * 45)
print("   TALLER AUTOMOTRIZ — LISTA DE VEHÍCULOS")
print("=" * 45)

for vehiculo in flota:
    vehiculo.mostrar_informacion()

total = sumar_total_mantenimiento(flota)

print("\n" + "=" * 45)
print(f"  COSTO TOTAL DE MANTENIMIENTO: ${total:,.2f}")
print("=" * 45)
```
